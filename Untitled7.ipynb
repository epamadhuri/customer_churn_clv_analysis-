```python
# Step 1: Convert transaction date to datetime
retail['trans_date'] = pd.to_datetime(retail['trans_date'], errors='coerce')

# Step 2: Define reference date (latest transaction in dataset)
ref_date = retail['trans_date'].max()

# Step 3: Build RFM features
rfm = retail.groupby('customer_id').agg(
    recency = ('trans_date', lambda x: (ref_date - x.max()).days),
    frequency = ('trans_date', 'count'),
    monetary = ('tran_amount', 'sum')
).reset_index()

print(rfm.head())
```

```python
# Step 1: Inspect churn dataset
print("Churn columns:", churn.columns.tolist())

# Step 2: Detect ID column
id_candidates = [c for c in churn.columns if 'id' in c.lower()]
print("Possible ID columns:", id_candidates)

# If we found one, rename it to customer_id
if len(id_candidates) > 0:
    churn = churn.rename(columns={id_candidates[0]:'customer_id'})
else:
    raise ValueError("No ID column found in churn dataset")

# Step 3: Detect churn label column
label_candidates = [c for c in churn.columns if 'churn' in c.lower()]
print("Possible churn label columns:", label_candidates)

if len(label_candidates) > 0:
    churn_label = label_candidates[0]
else:
    raise ValueError("No churn label column found in churn dataset")

# Step 4: Align types
features['customer_id'] = features['customer_id'].astype(str)
churn['customer_id'] = churn['customer_id'].astype(str)

# Step 5: Merge
data = features.merge(churn, on='customer_id', how='left')
print("Merged columns:", data.columns.tolist())

# Step 6: Define X and y automatically
X = data.drop(columns=['customer_id', churn_label])
y = data[churn_label]

print("Feature shape:", X.shape)
print("Target distribution:", y.value_counts())
```

```python
# Step 1: Encode categorical columns
categorical_cols = ['gender','subscription_type','contract_length']
X_encoded = pd.get_dummies(X, columns=categorical_cols, drop_first=True)

# Step 2: Train/Test Split again
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_encoded, y, test_size=0.2, random_state=42, stratify=y
)

# Step 3: Train Random Forest
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix

model = RandomForestClassifier(
    n_estimators=200,
    max_depth=10,
    random_state=42
)
model.fit(X_train, y_train)

# Step 4: Evaluate
y_pred = model.predict(X_test)
print("Classification Report:\n", classification_report(y_test, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
from xgboost import XGBRegressor

# Step 1: Target transformation
churn['log_spend'] = np.log1p(churn['total_spend'])

# Step 2: Feature engineering
churn['usage_per_month'] = churn['usage_frequency'] / churn['tenure']
churn['support_per_month'] = churn['support_calls'] / churn['tenure']

# Step 3: Define X and y
X_clv = churn.drop(columns=['customer_id','total_spend','log_spend','churn'])
y_clv = churn['log_spend']

# One-hot encode categorical features
X_clv = pd.get_dummies(X_clv, columns=['gender','subscription_type','contract_length'], drop_first=True)

# Step 4: Train/Test Split
X_train, X_test, y_train, y_test = train_test_split(
    X_clv, y_clv, test_size=0.2, random_state=42
)

# Step 5: Train XGBoost Regressor
xgb_reg = XGBRegressor(
    n_estimators=300,
    learning_rate=0.05,
    max_depth=6,
    subsample=0.8,
    colsample_bytree=0.8,
    random_state=42
)
xgb_reg.fit(X_train, y_train)

# Step 6: Evaluate
y_pred = xgb_reg.predict(X_test)
print("MSE:", mean_squared_error(y_test, y_pred))
print("R2:", r2_score(y_test, y_pred))

# Step 7: Feature importance visualization
importance = pd.DataFrame({
    'feature': X_train.columns,
    'importance': xgb_reg.feature_importances_
}).sort_values(by='importance', ascending=False)

plt.figure(figsize=(10,6))
plt.barh(importance['feature'][:10], importance['importance'][:10])
plt.gca().invert_yaxis()
plt.title("Top 10 Features Driving CLV (log spend)")
plt.show()
```

```python
# Churn probabilities
churn_probs = model.predict_proba(X_encoded)[:,1]  # probability of churn

# CLV predictions (log spend back-transformed)
clv_preds = np.expm1(xgb_reg.predict(X_clv))

# Combine into a DataFrame
results = pd.DataFrame({
    'customer_id': churn['customer_id'],
    'churn_prob': churn_probs,
    'predicted_spend': clv_preds
})

print(results.head())
```

```python
plt.figure(figsize=(8,6))
plt.scatter(results['predicted_spend'], results['churn_prob'], alpha=0.5)
plt.xlabel("Predicted Spend (CLV)")
plt.ylabel("Churn Probability")
plt.title("Customer Value vs Churn Risk")
plt.show()
```

```python
# Define thresholds (you can adjust these)
high_value = results['predicted_spend'].median()
high_risk = 0.5  # churn probability > 50%

results['segment'] = results.apply(
    lambda row:
        'High Value - High Risk' if row['predicted_spend'] > high_value and row['churn_prob'] > high_risk else
        'High Value - Low Risk' if row['predicted_spend'] > high_value else
        'Low Value - High Risk' if row['churn_prob'] > high_risk else
        'Low Value - Low Risk',
    axis=1
)

print(results['segment'].value_counts())
```

```python
sns.scatterplot(
    x='predicted_spend', y='churn_prob',
    hue='segment', data=results, alpha=0.6
)
plt.title("Customer Segmentation: Value vs Churn Risk")
plt.show()
```

```python
# Step 9: Interpret segments

# Count how many customers fall into each segment
segment_counts = results['segment'].value_counts()
print("Customer Segment Counts:\n", segment_counts)

# Summarize insights for each segment
insights = {
    'High Value - High Risk': "Critical group. These customers spend a lot but are likely to churn. Prioritize retention campaigns (loyalty rewards, personalized offers).",
    'High Value - Low Risk': "VIPs. They spend a lot and are loyal. Focus on upselling, cross-selling, and maintaining satisfaction.",
    'Low Value - High Risk': "Less urgent. They are likely to churn but contribute little revenue. Monitor for churn drivers but limit retention spend.",
    'Low Value - Low Risk': "Stable base. Low spend and low churn risk. Maintain cost-effective engagement."
}

print("\nBusiness Insights & Recommendations:")
for segment, message in insights.items():
    print(f"- {segment}: {message}")
```
